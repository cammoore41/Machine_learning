```python
import pandas as pd
import numpy as np
import sklearn.linear_model
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_squared_error
```

```python
test=pd.read_csv('../input/digit-recognizer/test.csv', header=0)
train=pd.read_csv('../input/digit-recognizer/train.csv', header=0)
```

```python
train.info()
```

```python
test.info()
```

```python
test.describe()
```

```python
train_y=train.iloc[:,0]
train_x=train.iloc[:,1:27999]
```

```python
train_y.describe()
```

```python
train_x.describe()
```

```python
test.describe()
```

```python
forest = RandomForestClassifier(n_estimators=100, random_state=1, n_jobs=-1)
%time forest.fit(train_x, train_y)
```

```python
pred1=forest.predict(test)
```

```python
type(pred1)
```

```python
np.savetxt('pred1.csv', pred1, delimiter=',')
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
x=np.concatenate((train_x,test),axis=0)
x=StandardScaler().fit_transform(x)
```

```python
pca=PCA(.95)
pca.fit(x)
totimages=pca.transform(x)
%time pca.n_components_
```

```python
trainnew=totimages[0:42000,:]
testnew=totimages[42000:70000,:]
```

```python
trainnew=trainnew.astype(int)
testnew=testnew.astype(int)
```

```python
%time forest.fit(trainnew, train_y)
```

```python
pred2=forest.predict(testnew)
```

```python
np.savetxt('pred2.csv', pred2, delimiter=',')
```
